```python
proj=project.drop(['Suggestions'],axis=1)
proj.head()
```

```python
proj["Stress"].value_counts()
```

```python
proj['Stress'].replace(['Low', 'Same','Medium','High'], [0, 1,1,2], inplace=True)
proj['Teaching_Level'].replace(['Lower Primary', 'Upper Primary', 'High School', 'Plus 2'], [0, 1, 2, 3], inplace=True)
proj['Prepration_Level'].replace(['Not sure', 'Less', 'Same', 'More'], [0, 1, 2, 3], inplace=True)
proj['Technical Expertise'].replace(['Low', 'Medium', 'satisfactory', 'High'], [0, 1,1,3], inplace=True)
proj['Online Teaching Effect'].replace(['Not sure','Less', 'Same', 'More'], [0, 1, 2, 3], inplace=True)
proj['Student_Attitude'].replace(['Poor','Good', 'Very good', 'Excellent'], [0, 1, 2, 3], inplace=True)
proj['Exam Performance'].replace(['poor','good', 'very good', 'Excellent'], [0, 1, 2, 3], inplace=True)
proj['Student Attentiveness'].replace(['poor','Satisfactory','Good', 'Very good'], [0, 1, 2, 3], inplace=True)
proj['Satisfaction'].replace(['Low', 'same','Medium','High'], [0,1, 1,2], inplace=True)
proj.head(15)
```

```python
proj.tail(50)
```

```python
final_data = pd.get_dummies(proj, columns = ['Gender','School_Category','State','Board','Platform'])
final_data.head()
```

```python
final_data.columns
```

```python
# Increase the size of the heatmap.
plt.figure(figsize=(20, 10))
# Store heatmap object in a variable to easily access it when you want to include more features (such as title).
# Set the range of values to be displayed on the colormap from -1 to 1, and set the annotation to True to display the correlation values on the heatmap.
heatmap = sns.heatmap(final_data.corr(), vmin=-1, vmax=1, annot=True)
# Give a title to the heatmap. Pad defines the distance of the title from the top of the heatmap.
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':12}, pad=12);
```

```python
from sklearn.model_selection import train_test_split
y = final_data[['Stress']]
X=final_data.drop(['Stress'],axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.05, random_state=0)
```

```python
from sklearn.metrics import confusion_matrix,roc_curve,auc,precision_score,recall_score,accuracy_score,f1_score
from sklearn.metrics import classification_report
```

```python
from sklearn.tree import DecisionTreeClassifier
dt_model = DecisionTreeClassifier(max_depth=4,random_state = 0)
dt_model.fit(X_train, y_train)
dt_pred = dt_model.predict(X_test)
dt_cm = confusion_matrix(y_test, dt_pred)
print ("Confusion Matrix : \n", dt_cm)
```

```python
from sklearn import tree
tree.plot_tree(dt_model)
```

```python
print("Evaluation of Decision Tree")
print("")
print(classification_report(y_test,dt_pred))
```

```python
dt_acc=accuracy_score(y_test, dt_model.predict(X_test))
print(dt_acc)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn_model = KNeighborsClassifier(n_neighbors = 1)
knn_model.fit(X_train, y_train)
knn_pred = knn_model.predict(X_test)
knn_cm = confusion_matrix(y_test, knn_pred)
print ("Confusion Matrix : \n", knn_cm)
```

```python
print("Evaluation of KNN")
print("")
print(classification_report(y_test,knn_pred))
```

```python
knn_acc=accuracy_score(y_test, knn_model.predict(X_test))
print(knn_acc)
```

```python
from sklearn.svm import SVC
svc_model = SVC()
svc_model.fit(X_train, y_train)
svm_pred = svc_model.predict(X_test)
svm_cm = confusion_matrix(y_test, svm_pred)
print ("Confusion Matrix : \n", svm_cm)
```

```python
svm_acc=accuracy_score(y_test, svc_model.predict(X_test))
print(svm_acc)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_model=RandomForestClassifier(n_estimators=100)
rf_model.fit(X_train,y_train)
rf_pred=rf_model.predict(X_test)
rf_cm = confusion_matrix(y_test, rf_pred)
print ("Confusion Matrix : \n", rf_cm)
```

```python
rf_acc=accuracy_score(y_test, rf_model.predict(X_test))
print(rf_acc)
```

```python
from sklearn.naive_bayes import GaussianNB
nb_model = GaussianNB()
nb_model.fit(X_train , y_train)
nb_pred = nb_model.predict(X_test)
```

```python
nb_acc=accuracy_score(y_test, nb_model.predict(X_test))
print(nb_acc)
```

```python

data = {'Model': ['Decision Tree', 'KNN', 'Naive Bayes', 'SVM','Random Forest'], 'Accuracy': [dt_acc,knn_acc,nb_acc,svm_acc,rf_acc]}  
  
# Create DataFrame  
df = pd.DataFrame(data)  
  
# Print the output.  
print(df)  
```
